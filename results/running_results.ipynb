You can use the following code to convert the running results in txt format to csv format.

```python
import pandas as pd
import re

input_file = 'input-your-log-file-here.txt'
output_file = 'output.csv'

data = []

with open(input_file, 'r') as file:
    for line in file:
        match = re.match(r'Game (\d+), Score (\d+), Record: (\d+), Time: ([\d.]+)s', line)
        if match:
            game, score, record, time = match.groups()
            data.append({'Game': game, 'Score': score, 'Record': record, 'Time': time})

df = pd.DataFrame(data)

df.to_csv(output_file, index=False)

print(f'Data has been processed and saved to {output_file}')
```

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
input_file = 'snake-ai-pytorch-dqn-data.csv'

df = pd.read_csv(input_file)

df['Average_Score'] = df['Score'].expanding().mean()

plt.figure(figsize=(10, 6))

plt.plot(df['Game'], df['Score'], label='Score', color='blue')

plt.plot(df['Game'], df['Average_Score'], label='Average Score', color='orange')

plt.plot(df['Game'], df['Record'], label='Record', color='green')

plt.title('Snake Agent - Performance with DQN')
plt.xlabel('Number of Games')
plt.ylabel('Score')
plt.legend()

plt.text(df['Game'].iloc[-1], df['Average_Score'].iloc[-1], f'{df["Average_Score"].iloc[-1]:.1f}', color='orange', ha='left')
plt.text(df['Game'].iloc[-1], df['Record'].iloc[-1], str(df['Record'].iloc[-1]), color='green', ha='left')

plt.grid(True)
plt.show()
````
